```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

X = df["Symptoms"]       
y = df["Disease"] 

vectorizer = CountVectorizer()
X_vectorized = vectorizer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))

def predict_disease(user_symptoms):

    user_vector = vectorizer.transform([user_symptoms])
    prediction = model.predict(user_vector)[0]

    remedy = df[df["Disease"] == prediction]["Medicine/Remedy"].values[0]
    return prediction, remedy

user_input = input("Enter your symptoms (separated by space): ")
disease, remedy = predict_disease(user_input)
print(f"\nPredicted Disease: {disease}")
print(f"Suggested Medicine/Remedy: {remedy}")
```

```python
import pickle

# Save model
pickle.dump(model, open('disease_nb_model.pkl', 'wb'))

# Save vectorizer
pickle.dump(vectorizer, open('vectorizer.pkl', 'wb'))

print("✅ Model and Vectorizer saved successfully!")

```
